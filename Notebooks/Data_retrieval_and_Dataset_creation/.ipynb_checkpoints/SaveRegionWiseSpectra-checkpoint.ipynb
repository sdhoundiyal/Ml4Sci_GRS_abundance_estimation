```python
#set the directory containing the .dat and .xml files
gammaRayDataDirectory="D:/Non-academic/GSOC23/Dataset"
```

```python
#get the files in the directory
allFilesInGRDataDirectory=os.listdir(gammaRayDataDirectory)
#create a list containing only the xml files
xmlFilesAddresses=[]
#iterate through all the files
for currentFile in allFilesInGRDataDirectory:
    #check if the file is an .xml file
    if currentFile[-3:]=='xml':
        xmlFilesAddresses.append(gammaRayDataDirectory+'/'+currentFile)
```

```python
#create empty numpy arrays to hold:
#the accepted spectrum
#the deadtime
#the altitude
#the latitude
#the longitude
acceptedSpectra=np.array([[0]*512])
deadtimes=np.array([])
altitudes=np.array([])
latitudes=np.array([])
longitudes=np.array([])
```

```python
#iterate over all the xml files
for currentXMLFile in xmlFilesAddresses:
    #read the XML file into a table
    dataTable=pds4_tools.read(currentXMLFile)[0]
    #read the data from the table
    currentFileAcceptedSpectra=dataTable.fields[0]
    currentFileDeadTimes=dataTable.fields[2]
    currentFileAltitudes=dataTable.fields[6] 
    currentFileLatitudes=dataTable.fields[7] 
    currentFileLongitudes=dataTable.fields[8] 
         
    acceptedSpectra=np.concatenate((acceptedSpectra,
                                    currentFileAcceptedSpectra)) 
    deadtimes=np.concatenate((deadtimes,
                              currentFileDeadTimes))
    altitudes=np.concatenate((altitudes,
                              currentFileAltitudes))      
    latitudes=np.concatenate((latitudes,
                              currentFileLatitudes))      
    longitudes=np.concatenate((longitudes,
                               currentFileLongitudes))  
```

```python
#remove the 0th, which contains all 0s and was added for computational convenience
acceptedSpectra=acceptedSpectra[1:,:]
```

```python
#print the shapes of the retrieved data
print("Shapes of the data are:")
print(f"Accepted spectra:\t{acceptedSpectra.shape}")
print(f"Dead times:\t\t{deadtimes.shape}")
print(f"Altitudes:\t\t{altitudes.shape}")
print(f"Latitudes:\t\t{latitudes.shape}")
print(f"Longitudes:\t\t{longitudes.shape}")
```

```python
#set the address of the the file containing oxygen abundances for 5 degree regions
oxygenAbundanceFileAddress="D:/Non-academic/GSOC23/ElementalAbundances/oxygen5d.txt"
```

```python
#open the file in read mode
currentFile=open(oxygenAbundanceFileAddress,"r")
#read each line from the file into a list
oxygenAbundanceLines=currentFile.readlines()
#close the file
currentFile.close()
```

```python
#iterate over all the lines and print them and the line number
for i,currentLine in zip(range(len(oxygenAbundanceLines)),oxygenAbundanceLines):
    print(i,currentLine)
```

```python
#remove the first 13 lines (indices with 0-12) as they contain metadata
oxygenAbundanceLines=oxygenAbundanceLines[13:]
```

```python
#import the regex package
import re
```

```python
#iterate over all the lines and print them and the line number
for i,currentLine in zip(range(len(oxygenAbundanceLines)),oxygenAbundanceLines):
    print(i,currentLine)
```

```python
#set the upper and lower limits for the desired altitudes
upperAltitudeLimit=100.0+10
lowerAltitudeLimit=100.0-10
```

```python
#adjust longitude values so that all values >180 are converted to -180 to 0
adjustedLongitudes=np.copy(longitudes)
adjustedLongitudes[adjustedLongitudes>180]=adjustedLongitudes[adjustedLongitudes>180]-360
```

```python
#create arrays to hold the minium and maximum latitudes and longitudes
maxLatitudePerSection=[]
minLatitudePerSection=[]
maxLongitudePerSection=[]
minLongitudePerSection=[]
#create an array to hold the number of spectra in each section
noOfSpectraPerSection=[]
#create an array to hold the active times
totalActiveTimePerSection=[]
#create an array to hold the summed spectra
summedSpectraPerSection=[]
#create an aray to hold the normalized spectra
normalizedSpectraPerSection=[]
```

```python
#iterate over all the lines and print them and the line number
for i,currentLine in zip(range(len(oxygenAbundanceLines)),oxygenAbundanceLines):
    #use the regex to extract each floating point number from the current line into a list containing:
    #Lat(min,max)   Lon(min,max)   O weight percent
    currentLineInfo=re.findall(r"[-+]?\d*\.\d+|\d+", currentLine)
    #retrieve the minimum and maximum latitude and longitudes
    minLatitude=float(currentLineInfo[0])
    maxLatitude=float(currentLineInfo[1])
    minLongitude=float(currentLineInfo[2])
    maxLongitude=float(currentLineInfo[3])
    
    
    #get indices of the accepted spectra collected lying withing the current latitudes and longitudes, and the desired altitude limits
    currentSectionSpectraIndices=np.where(np.logical_and(np.logical_and(np.logical_and(latitudes>=minLatitude,
                                                                                       latitudes<=maxLatitude),
                                                                        np.logical_and(adjustedLongitudes>=minLongitude,
                                                                                       adjustedLongitudes<=maxLongitude)),
                                                         np.logical_and(altitudes>=lowerAltitudeLimit,
                                                                        altitudes<=upperAltitudeLimit)))
    
    #save the maximum and minimum latitudes and longitudes as well as number of spectra in the section
    maxLatitudePerSection.append(maxLatitude)
    minLatitudePerSection.append(minLatitude)
    maxLongitudePerSection.append(maxLongitude)
    minLongitudePerSection.append(minLongitude)
    noOfSpectraPerSection.append(currentSectionSpectraIndices[0].shape[0])
    
    #get the spectra in the current region
    currentRegionSpectra=acceptedSpectra[currentSectionSpectraIndices,:][0,:,:]
    #get the deadtimes for the spectra in the current region
    currentRegionDeadtimes=deadtimes[currentSectionSpectraIndices]
    #get the total active time during which the spectra were collected
    totalActiveTime=np.sum(32*(1-currentRegionDeadtimes))
    #get the summed spectra
    currentSummedSpectra=np.sum(currentRegionSpectra,axis=0)
    #normalize the spectra to counts per minute
    currentNormalizedSpectra=(currentSummedSpectra/totalActiveTime)*60
    
    #save the total active time, summed spectra, and normalized spectra
    totalActiveTimePerSection.append(totalActiveTime)
    summedSpectraPerSection.append(currentSummedSpectra)
    normalizedSpectraPerSection.append(currentNormalizedSpectra)
    
    print(totalActiveTime)
                                          
    #iterate over all the spectra in the current region and correct them for the time
```

```python
#convert the list to numpy array
maxLatitudePerSection=np.array(maxLatitudePerSection)
minLatitudePerSection=np.array(minLatitudePerSection)
maxLongitudePerSection=np.array(maxLongitudePerSection)
minLongitudePerSection=np.array(minLongitudePerSection)
noOfSpectraPerSection=np.array(noOfSpectraPerSection)
totalActiveTimePerSection=np.array(totalActiveTimePerSection)
summedSpectraPerSection=np.vstack(summedSpectraPerSection)
normalizedSpectraPerSection=np.vstack(normalizedSpectraPerSection)
```

```python
#compute the log of the normalized spectra
logScaledNormalizedSpectraPerSection=np.log10(normalizedSpectraPerSection)
```
