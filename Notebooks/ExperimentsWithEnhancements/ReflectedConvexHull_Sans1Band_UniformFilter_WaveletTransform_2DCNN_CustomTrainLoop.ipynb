```python
#import the uniform filter from scipy
from scipy.signal import savgol_filter
#import the function for convex hull removal from pysptools
from pysptools.spectro import convex_hull_removal
```

```python
#define a function which given GR spectra and the size of a filter, performs mean filtering on the GRS spectra
def applySavgolFilterToSpectra(spectra,filterSize):
    #filter the spectra
    filteredSpectra=savgol_filter(spectra,
                                  window_length=filterSize,
                                  polyorder=2,
                                  axis=-1,
                                  mode='mirror')
    #return the filtered spectra
    return filteredSpectra
```

```python
#define a function which takes GR spectra and performs continuum removal on them
def removeGRContinuum(spectra):
    #scale the spectra between 0 and 1
    spectra=(spectra-np.amin(spectra,axis=-1,keepdims=True))/np.ptp(spectra,axis=-1,keepdims=True)
    
    #reflect the spectra along the line, y=max counts
    reflectedSpectra=np.amax(spectra)+(np.amax(spectra)-spectra)
    
    #create an array to hold the continuum removed spectra
    continuumRemovedSpecta=np.zeros_like(reflectedSpectra)
    
    #iterate through the reflected spectra and remove the continuum
    for currentSpectraIndex in range(reflectedSpectra.shape[0]):
        continuumRemovedSpecta[currentSpectraIndex,:]=convex_hull_removal(reflectedSpectra[currentSpectraIndex,:],
                                                                          energyBands)[0]
        
    #reflect the continuum removed spectra along the line, y=max counts
    reflectedContinuumRemovedSpecta=(np.amax(continuumRemovedSpecta)-continuumRemovedSpecta)
    
    #return the continuum removed spectra
    return reflectedContinuumRemovedSpecta
```

```python
#define a function which takes GR spectra and normalized them (i.e. makes their l2-norm equal to one)
def normalizeSpectra(spectra):
    #create an array to hold the normalized spectra
    normalizedSpectra=np.zeros_like(spectra)
    
    #iterate through the spectra and normalize each
    for i in range(spectra.shape[0]):
        normalizedSpectra[i,:]=spectra[i,:]/np.linalg.norm(spectra[i,:])
        
    #return the normalized spectra
    return normalizedSpectra
```

```python
#define a function which given an array of super-pixels, explained-variance ratio, and no of super-pixels to be selected
#returns the indices of the interesting super-pixels in the array
def applyDEMUDToSpectra(spectra,varianceToExplain,noOfSpectraToRetrive):
    #create an array to hold the indices of the selected (interesting) super-pixels
    selectedSpectraIndices=[]
    #create a PCA object which explains over 95% of the variance in the data
    pcaObject=PCA(n_components=varianceToExplain,svd_solver='auto')
    #compute the PCA model for the super-pixels and get the low dimensional representation, and reconstruct the orginal spectra
    reconstructedspectra=pcaObject.inverse_transform(pcaObject.fit_transform(spectra))
    #create an array to save the reconstruction error (cosine distance)
    reconstructionErrors=np.full(spectra.shape[0],np.nan)
    #compute the cosine distance between the orginal and reconstructed spectra
    for i in range(spectra.shape[0]):
        reconstructionErrors[i]=cosineDistance(spectra[i,:],reconstructedspectra[i,:])
    #save the index of the spectra with the greatest error
    selectedSpectraIndices.append(np.argmax(reconstructionErrors))

    #iteratively extract the most dissimar spectra
    for i in range(noOfSpectraToRetrive-1):
        #fit PCA to the selected spectra
        pcaObject.fit(spectra[selectedSpectraIndices,:])
        #apply PCA and then reconstruct the spectra
        reconstructedspectra=pcaObject.inverse_transform(pcaObject.transform(spectra))
        #create an array to save the reconstruction error (cosine distance)
        reconstructionErrors=np.full(spectra.shape[0],np.nan)
        #compute the cosine distance between the orginal and reconstructed spectra
        for i in range(spectra.shape[0]):
            reconstructionErrors[i]=cosineDistance(spectra[i,:],reconstructedspectra[i,:])
        #remove the error of the selected pixels
        reconstructionErrors=np.delete(reconstructionErrors,selectedSpectraIndices)
        #get the index of the spectra with the greatest error
        selectedSpectraIndices.append(np.delete(np.arange(0,spectra.shape[0],1),selectedSpectraIndices,axis=0)[np.argmax(reconstructionErrors)])
        
    return selectedSpectraIndices
```

```python
#define a function which computes the gradient for one batch of training data 
def computeGradientForOneBatch(currentBatchInputValues,currentBatchOutputValues):
    #get the model
    global abundanceEstimator
    #get the optimizer
    global optimizer
    
    #compute the gradient and use it to update the wieghts
    with tf.GradientTape() as tape:

        #compute the output of the network
        predictedOutputValues=abundanceEstimator(currentBatchInputValues)

        #compute the loss
        loss=lossCalculator(tf.cast(predictedOutputValues,
                                     tf.float32),
                             tf.cast(currentBatchOutputValues,
                                     tf.float32))

        #compute the gradient for the total loss
        gradientValues=tape.gradient(loss,
                                     abundanceEstimator.trainable_variables)
    
    
    #return the gradient
    return loss,gradientValues
```

```python
#define a function which given training data fits the model to it for a single epoch
def fitModelToDataForOneEpoch(inputValues,outputValues):
    #get the model
    global abundanceEstimator
    #get the batch size
    global batchSize
    
    #iterate over all samples
    for i in range(0,inputValues.shape[0],batchSize):
        #get the training spectra and abundances in the current batch
        currentBatchSpectra=inputValues[i:i+batchSize,:]
        currentBatchAbundances=outputValues[i:i+batchSize]

        #compute the gradient for the total loss
        loss,gradientValues=computeGradientForOneBatch(currentBatchSpectra,currentBatchAbundances)

        #optimize for the loss
        optimizer.apply_gradients(zip(gradientValues,abundanceEstimator.trainable_variables))
    
    #return the loss from the last batch of the current epoch
    return loss
```

```python
#define a funtion which given input and output values for testing data computes the loss
def computeLoss(inputValues,outputValues):
    #get the model
    global abundanceEstimator

    #compute the output of the network
    predictedOutputValues=abundanceEstimator(inputValues)
    #compute the loss
    loss=lossCalculator(tf.cast(outputValues,
                                 tf.float32),
                         tf.cast(predictedOutputValues,
                                 tf.float32))
    
    #return the loss
    return loss
```

```python
#define a function for computing R2
def comuputePearsonsCorrelation(inputValues,outputValues):
    #get the model
    global abundanceEstimator
    
    #compute the output of the network
    predictedOutputValues=abundanceEstimator(inputValues).numpy()[:,0]
    
    
    #comupte and the r2 score
    return np.corrcoef(np.array([predictedOutputValues,outputValues+1e-12]))[0,1]
```

```python
#define a function which takes training and test data, and no of epochs and fits a model to the data
def fitModelToData(trainingInputs,trainingOutputs,testingInputs,testingOutputs,noOfEpochs,untrainedModelAddress):
    #build the model

    global abundanceEstimator
    global batchSize
    abundanceEstimator=tf.keras.models.load_model(untrainedModelAddress,
                                                  compile=False)
    #create lists to hold training and test losses
    trainingLosses=[]
    testingLosses=[]
    trainingCorrelationValues=[]
    testingCorrelationValues=[]
    
    
    #train the network for the specified number of epochs
    for currentEpoch in range(noOfEpochs):

        #fit the model to the data and compute the loss
        currentTrainingLoss=fitModelToDataForOneEpoch(trainingInputs,
                                                      trainingOutputs).numpy()

        #save the traininig loss 
        trainingLosses.append(currentTrainingLoss)



        #compute the testing loss
        currentTestLoss=0
        for i in range(0,testingInputs.shape[0],batchSize):
            currentTestLoss+=computeLoss(testingInputs[i:i+batchSize,:,:],
                                         testingOutputs[i:i+batchSize]).numpy()

        #save the loss for the last batch
        testingLosses.append(currentTestLoss)


        #compute and save the corr values for the training and test data
        currentTrainingCorrelationValue=0
        for i in range(0,testingInputs.shape[0],batchSize):
            currentTrainingCorrelationValue=comuputePearsonsCorrelation(trainingInputs[i:i+batchSize,:,:],
                                                                         trainingOutputs[i:i+batchSize])
            
        trainingCorrelationValues.append(currentTrainingCorrelationValue)
        
        currentTestingCorrelationValue=0
        for i in range(0,testingInputs.shape[0],batchSize):
            currentTestingCorrelationValue=comuputePearsonsCorrelation(testingInputs[i:i+batchSize,:,:],
                                                                         testingOutputs[i:i+batchSize])

        testingCorrelationValues.append(currentTestingCorrelationValue)
        
        #print progress
        if currentEpoch%(noOfEpochs//20)==0:
            print(f"{currentEpoch}:\t{round(trainingCorrelationValues[-1],2)}\t{round(testingCorrelationValues[-1],2)}")
            
            
    #delete the model from memory
    abundanceEstimator=None
    
    
    #combine the learning curves into a dictionary
    learningCurves={'Train Loss':trainingLosses,
                    'Test Loss':testingLosses,
                    'Train Acc.':trainingCorrelationValues,
                    'Test Acc.':testingCorrelationValues}
    
    #return the learning curves
    return learningCurves
```

```python
#define a function which takes input and output data and number of folds and fits a cross-validated model to the data
def fitCrossValidatedModel(inputData,outputData,noOfFolds,noOfEpochs,untrainedModelAddress):
    #get the seed
    global globalSeed
    #create a K-Fold splitter
    splitter=KFold(n_splits=noOfFolds,
                   shuffle=True,
                   random_state=globalSeed)
    
    #create an empty dictionary to hold the learning curves for all folds
    foldWiseLearningCurves={}
    
    #iterate through the folds
    for i,(trainingIndices,testingIndices) in enumerate(splitter.split(outputData)):
        
        #print the number of the current fold
        print(f"Fold no.:\t{i}")
        
        #get the training and testing spectra
        trainingInputData=inputData[trainingIndices,:]
        testingInputData=inputData[testingIndices,:]

        #get the training and testing abundances
        trainingOutputData=outputData[trainingIndices]
        testingOutputData=outputData[testingIndices]
        
        #train a model for the current fold and get its learning curves
        currentFoldLearningCurve=fitModelToData(trainingInputData,
                                                trainingOutputData,
                                                testingInputData,
                                                testingOutputData,
                                                noOfEpochs,
                                                untrainedModelAddress)
        
        #append the current fold's learning curve to the dictionary
        foldWiseLearningCurves[(i+1)]=currentFoldLearningCurve
        
    
    #return the fold-wise learning curves
    return foldWiseLearningCurves
```

```python
#define a function which uses the given spectra to build models for predicting abundances for all elements
#and return the element wise learning curves
def predictElementWiseAbundances(spectra,noOfFolds,noOfEpochs,untrainedModelAddress):
    
    #get access to the element names and abundances
    global elementNames
    global currentMineralAbundancevalues

    #create a dictionary to hold element-wise learning curves
    elementWiseLearningCurves={}
    

    #iterate through all the elements
    for currentMineralIndex,currentElementName in zip(range(len(elementNames)),elementNames):
        print(f"{currentMineralIndex+1}---{currentElementName}")

        #select the mineral whose abundance is to be estimated
        currentMineralAbundancevalues=abundanceValues[currentMineralIndex]
        #scale the abundance values between 0 and 1
        scaledAbundanceValuesCurrentAbundanceValues=(currentMineralAbundancevalues-np.amin(currentMineralAbundancevalues))/np.ptp(currentMineralAbundancevalues)
        
        #get learning curves for the current element
        currentElementLearningCurves=fitCrossValidatedModel(spectra,
                                                            scaledAbundanceValuesCurrentAbundanceValues,
                                                            noOfFolds,
                                                            noOfEpochs,
                                                            untrainedModelAddress)
        
        #save the learning's curves
        elementWiseLearningCurves[currentElementName]=currentElementLearningCurves
    
    
    #return the learning curves
    return elementWiseLearningCurves
```
